Problem 3: Given two compatible matrices, write a Python function that performs Matrix-Matrix
product. Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-
Matrix product.

```python
def matrix_matrix_product(A,B):
    res=np.zeros((len(A),len(B[0])))
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                res[i][j]+= A[i][k]*B[k][j]
    return res
A = np.array([[2,4],[4,5]])
B = np.array([[5,9],[3,8]])
print(matrix_matrix_product(A, B))
print(np.dot(A, B))
```

Problem 4: Given a matrix, write a Python function that checks if it is a symmetric matrix.

```python
def is_symmetric(A):
    return np.array_equal(A,A.T)
A=np.array([[1,2,3],[2,5,6],[3,6,9]])
print(is_symmetric(A))
```
