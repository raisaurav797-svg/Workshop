The following table shows the different vehicles and the distance covered by them in the span of 5 days. Display the following table in line plot.  
a.	Plot “days” in X- axis and “distance covered” in Y- axis.
b.	Title of the table “bike details in Plot”



```python
import matplotlib.pyplot as plt
days=[1, 2, 3, 4, 5]
bike1=[100, 120, 130, 140, 150]
bike2=[90, 110, 115, 125, 135]
bike3=[80, 95, 105, 115, 120]
bike4=[70, 85, 95, 105, 110]
plt.plot(days, bike1, marker='o', label='Bike 1', linewidth=2)
plt.plot(days, bike2, marker='s', label='Bike 2', linewidth=2)
plt.plot(days, bike3, marker='^', label='Bike 3', linewidth=2)
plt.plot(days, bike4, marker='D', label='Bike 4', linewidth=2)
plt.xlabel('Days')
plt.ylabel('Distance Covered (km)')
plt.legend()
plt.grid()
plt.show()
```

9.	Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋], i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋

```python
import matplotlib.pyplot as plt
import math
x=[]
y=[]
i =-3 * math.pi
while i <= 3 * math.pi:
    x.append(i)
    y.append(math.cos(i))
    i += 0.1
plt.plot(x, y, color='blue')
plt.grid()
plt.show()

```

10.	Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
import matplotlib.pyplot as plt
r=4
center=[2,2]
x=[]
y=[]
theta=0.0
while(theta <= 2* math.pi):
    x.append(r*math.cos(theta)+center[0])
    y.append(r*math.sin(theta)+center[1])
    theta += 0.125
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.plot(x, y, '-')
plt.grid()
plt.show()
```
