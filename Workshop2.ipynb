1. Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def work(i):
    unique=[]
    for item in i:
        if item not in unique:
            unique.append(item)
    return unique
sample= [1,2,3,3,3,3,4,5]
print("Sample List :",sample)
print("Unique List :",work(sample))
```

2. Write a Python function that takes a number as a parameter and checks whether the number
is prime or not.

```python
def prime(n):
    if n<2:
        return False
    if n==2:
        return True
    if n%2==0:
        return False
    for i in range(3,int(n **0.5) +1, 2):
        if n % i == 0:
            return False
    return True
print({prime(6)})
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
def primes(limit):
    print("Twin Primes less than 1000:")
    twins=[]
    for n in range(3, limit - 1): # need n+2 < limit
        if prime(n) and prime(n + 2):
            twins.append((n, n + 2))
    for pair in twins:
        print(pair)
print(primes(1000))
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sum_of_divisors(n):
    #Returns the sum of all divisors of n, excluding n itself
    total=0
    for i in range(1, n):
        if n%i==0:
            total+= i
    return total
print(sum_of_divisors(14))
```

6. Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
def mean(marks):
    total=0
    for mark in marks:
        total+=mark
    return total/len(marks)
def median(marks):
    sorted_marks=sorted(marks)
    n=len(sorted_marks)
    mid=n//2
    if n % 2 == 0:
        return (sorted_marks[mid - 1] + sorted_marks[mid]) / 2
    else:
        return sorted_marks[mid]
Marks=[20,30,35,55,70,80,90]
print(f"Mean:{mean(Marks)}")
print(f"Median:{median(Marks)}")
```

7. Write a function that converts a decimal number to binary number.

```python
def deci(n):
    if n==0:
        return "0"
    bin=""
    neg=n<0
    n=abs(n)
    while n>0:
        bin=str(n% 2)+bin
        n=n//2
    if neg:
        bin="-"+bin
    return bin
for num in [0,5,10,25,255]:
    print("The conversion of decimal number to binary number:")
    print(f"{num:>5}={deci(num)}")
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def pro(n):
    n=abs(n)
    product=1
    for digit in str(n):
        product*=int(digit)
    return product
for num in [234,999,105,7]:
    print("The product of digits of the numbers:")
    print(f"{num}={pro(num)}")
```

9. Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:
a. 1,2,3,4,5
b. 12,13,14

```python
def multi(*no):
    res=1
    for num in no:
        res*=num
    return res
print(f"The multiplication of (1,2,3,4,5):{multi(1,2,3,4,5)}")
print(f"The multipilcation of (12,13,14):{multi(12,13,14)}")
```
