```python
def myPow(x, n):
    res=1
    while n>0:
        res*=x
        n-=1
    return res
print(myPow(8,15))
```

```python
def SumDigits(n):
    n=abs(n)
    total=0
    while n>0:
        total+=n%10
        n//=10
    print("Sum of digits:",total)
SumDigits(234)
```

```python
def squaredSum(sq):
    total=0
    for num in sq:
        total+=num*num
    return total
print(squaredSum([8,3,1]))
```

```python
def sumDigits(num):
    total=0
    i=1
    while i<=num:
        total+=i
        i+=1
    return total
print(sumDigits(7))
```
