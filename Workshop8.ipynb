```python
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1 and 2

y1 = (5-x)/3
y2 = (5+x)/2
y3 = 3-x
## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(0.5,0.6, '$x+3y = 5$', color='r')

ax.text(-3,0.5,'$x-2y = -5$', color='g')

ax.text(1.2,2, '$x+y = 3$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-2,6)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

Task 1:

Plot the graph for linear system b and write the point of intersection between three lines.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y1 =(3-2*x)/3
y2 = (6-4*x)/6
y3 = (3*x-1)/2
fig, ax=plt.subplots()
ax.plot(x, y1, color='r', linewidth=2)
ax.plot(x, y2, color='g', linewidth=1, linestyle='--')
ax.plot(x, y3, color='b')
A = np.array([[2, 3], [3, -2]])
b_vec = np.array([3, 1])
intersection = np.linalg.solve(A, b_vec)
ax.plot(intersection[0], intersection[1], 'ko', markersize=8)
ax.annotate(f'  ({intersection[0]:.2f}, {intersection[1]:.2f})',xy=intersection, fontsize=9)
ax.text(1.5,  0.8, '$2x+3y=3$',  color='r')
ax.text(0.5, -1.2, '$4x+6y=6$',  color='g')
ax.text(-3.5, -1.9, '$3x-2y=1$', color='b')
ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)
ax.set_xticks(list(range(-4, 5)))
ax.set_aspect('equal')
ax.grid(True, ls=':')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear System - Part b")
plt.show()
```

Task 4:

1. Consider the following 10 points in X-Y plane: (−3, −7.6), (−2,1), (−1, 4.8), (0, 5), (1, 2.8), (2, −0.6), (3, −4), (4, −6.2), (5, −6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-3,-2,-1,0,1,2,3,4,5,7], dtype=float)
y = np.array([-7.6,1,4.8,5,2.8,-0.6,-4,-6.2,-6,6.4], dtype=float)

coeffs = np.linalg.solve(np.vander(x, N=10), y)
poly = np.poly1d(coeffs)

x_fit = np.linspace(-3, 7, 500)
plt.scatter(x, y, color='red')
plt.plot(x_fit, poly(x_fit))
plt.grid(True, alpha=0.3)
plt.show()

for xi, ya in zip(x, y):
    print(f"{xi:>3} | {poly(xi):>9.4f} | {ya:>6.1f} | {np.isclose(poly(xi), ya)}")
```
