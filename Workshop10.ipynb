Problem 1: Consider varying the value of step size ℎ to a bigger value
and smaller value and check for the error. Consider varying the value of
𝑎𝑎 and again check the error.


```python
import numpy as np

def derivative(func, a, method='forward', h=0.001):
    if method == 'forward':
        return (func(a + h) - func(a)) / h
    elif method == 'backward':
        return (func(a) - func(a - h)) / h
    else:
        raise ValueError("Method must be 'forward' or 'backward'")

# True derivative of cos(x) = -sin(x)
actual_value = -np.sin(0.0)
h_values = [1, 0.1, 0.01, 0.001, 0.0001]
print("Testing different h values:\n")
for h in h_values:
    result = derivative(np.cos, 0, method='forward', h=h)
    error = abs(result - actual_value)
    print(f"h = {h}")
    print(f"Approx Derivative = {result}")
    print(f"Error = {error}\n")
    a_values = [0, np.pi/4, np.pi/2, np.pi]

h = 0.001
print("Testing different a values:\n")
for a in a_values:
    actual = -np.sin(a)
    result = derivative(np.cos, a, method='forward', h=h)
    error = abs(result - actual)
    print(f"a = {a}")
    print(f"Actual Value = {actual}")
    print(f"Approx Value = {result}")
    print(f"Error = {error}\n")
```

Problem 2: In the above code, only the Forward and Backward
Difference formulae have been implemented. Please add the a few lines
of code to derivative() function to implement the Central Difference
formula and then compare its result with the Forward and Backward
Difference methods.

```python
import numpy as np
def derivative(func, a, method='forward', h=0.001):
    if method == 'forward':
        return (func(a + h) - func(a)) / h
    elif method == 'backward':
        return (func(a) - func(a - h)) / h
    elif method == 'central':
        return (func(a + h) - func(a - h)) / (2 * h)
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'")

a = 0
actual_value = -np.sin(a)
forward_result = derivative(np.cos, a, method='forward')
backward_result = derivative(np.cos, a, method='backward')
central_result = derivative(np.cos, a, method='central')
print("Actual Value =", actual_value)
print("\nForward Difference =", forward_result)
print("Forward Error =", abs(forward_result - actual_value))
print("\nBackward Difference =", backward_result)
print("Backward Error =", abs(backward_result - actual_value))
print("\nCentral Difference =", central_result)
print("Central Error =", abs(central_result - actual_value))
```

Problem 3: The derivative() function can take an array of inputs for 𝑎𝑎
and return the derivatives for each 𝑎𝑎 value. So, calculate the derivative of 𝑦𝑦 = 𝑠𝑠𝑙𝑙𝑠𝑠(𝑥𝑥) for multiple equidistant values and plot it along the true
values.

```python
import numpy as np
import matplotlib.pyplot as plt

def derivative(func, a, method='central', h=0.001):
    if method == 'forward':
        return (func(a + h) - func(a)) / h
    elif method == 'backward':
        return (func(a) - func(a - h)) / h
    elif method == 'central':
        return (func(a + h) - func(a - h)) / (2 * h)
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'")
x = np.linspace(0, 5*np.pi, 100)
y = np.sin(x)
dy_numerical = derivative(np.sin, x, method='central')
dy_actual = np.cos(x)
plt.plot(x, dy_actual, label='True Derivative cos(x)')
plt.plot(x, dy_numerical, '--', label='Numerical Derivative')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.title('Numerical vs True Derivative')
plt.legend()
plt.grid(True)
plt.show()
```
