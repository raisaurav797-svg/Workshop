Task-1:
Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;

What happens?

```python
import matplotlib.pyplot as plt
import random
x=[random.randint(0, 50) for _ in range(10)]
y=[random.randint(0, 50) for _ in range(10)]
colors=["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]
plt.figure(figsize=(7, 5))
plt.scatter(x, y,c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

Task 2:
Following data shows the change in population of different animals from 2017 to 2022.
a. Create a pie chart for the population of each year
b. Can we show all the pie chart on the single pane? Write the code.

```python
import matplotlib.pyplot as plt
year=[2017, 2018, 2019, 2020, 2021, 2022]
bear=[81,54,93,116,137,184]
dolphin=[25,77,32,72,62,34]
whale=[80,54,100,76,93,72]
animals=['bear','dolphin','whale']
fig,axes=plt.subplots(2, 3, figsize=(14, 9))
fig.suptitle("Animal Population 2017–2022", fontsize=14, fontweight='bold')
for i, (ax, year) in enumerate(zip(axes.flat,year)):
    data=[bear[i],dolphin[i],whale[i]]
    ax.pie(data, labels=animals)
    ax.set_title(str(year))
plt.show()
```

```python
import matplotlib.pyplot as plt
fruits=["Mango","Orange","Plum","Pineapple","Melon"]
counts=[45,30,15,30,30]
plt.pie(counts,labels=fruits)
plt.grid(color='#95a5a6', linestyle='--',linewidth=2, axis='y',alpha=0.7)
plt.show()
```

```python

import matplotlib.pyplot as plt
y=[123,230,78,193,12]
x=["Red","Silver","Blue","White","Green"]
car_colors=["red","silver","blue","yellow","green"]
plt.bar(x,y,width=0.3,color=car_colors)
#plt.barh(x,y)
plt.title("A bar plot")
plt.xlabel("Colors")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
y=[123,230,78,193,12]
x=["Red","Silver","Blue","White","Green"]
car_colors=["red","silver","blue","yellow","green"]
plt.pie(y,labels=x,colors=car_colors)
plt.title("Car Colours Pie Chart")
myexplode = [0.3, 0, 0, 0, 0]
mycolours = ["Magenta","Green","Cyan","Red","Orange"]
plt.pie(y,labels=x,explode=myexplode,shadow=True,colors=mycolours)
plt.legend(title = "Car Colours:")
plt.show()
```
